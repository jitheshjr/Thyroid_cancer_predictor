```python
# Gradient Descent for Linear Regression

import numpy as np
import matplotlib.pyplot as plt

def grad(x,y,learning_rate,epochs):
    m = 0
    b = 0
    n = len(x)
    costs = []

    for epoch in range(epochs):

        y_pred = m*x + b
        error = y_pred - y

        m -= learning_rate * (1/n) * np.sum(error*x)
        b -= learning_rate * (1/n) * np.sum(error)

        if epoch % 100 == 0:
            cost = (1 / (2*n) * np.sum(error**2))
            print(f"Epoch {epoch} :- cost = {cost}")
            costs.append(cost)

    plt.plot(range(0,epochs,100),costs,marker='o')
    plt.xlabel("Epoches")
    plt.ylabel("Cost")
    plt.grid(True)
    plt.title("Epoch vs Cost")
    plt.show()

    return m,b

x = np.random.rand(100)
y = (2*x + 3)*0.1
learning_rate = 0.1
epochs = 1000

m,b = grad(x,y,learning_rate,epochs)
print(f"Slope: {m}")
print(f"Intercept: {b}")
```
